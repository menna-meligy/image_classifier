```python
#preprocessing data

data = data.map(lambda x , y : (x/122.5 , y))
```

```python
# split data -> training , testing partitions to validate it through making sure that the model data hasn't overfit and got specific partitions 
# 1- establishing what a training data sizes should be
# len(data) 7
# train and val ->evaluate our model while training
# test partition, model won't be seen until the final stage of evaluation
# 4 , 2, 1
train_size = int(len(data)*.7) 
val_size = int(len(data)*.2)+1
test_size = int(len(data)*.1)+1
```

```python
# using take and skip in ts dataset pipeline
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size + val_size).take(test_size)
# stablising the 3 stages partitions 
```

```python
# building the deep learning -> keras sequential api
# 1- build deep learning model ->train .fit method->plot it using mathplot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model = Sequential()
```

```python
# Conv2D(12 , (3,3) , 1 ->#of filters , size of the filter and the stride
# relu activation ->converting any - to 0 and any + becomes unchanged , another option -> segmoid activation
model.add(Conv2D(16 , (3,3) , 1 , activation='relu' , input_shape=(256,256,3)))
# maximizing value after the relu activation and return it -> condensation
model.add(MaxPooling2D())

model.add(Conv2D(32 , (3,3) , 1 , activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16 , (3,3) , 1 , activation='relu'))
model.add(MaxPooling2D())
# flatting before densing cuz -> it is flattining the width and the rows and # of filters will form the channels value 
# so it will return 1 value for densing
model.add(Flatten())

model.add(Dense(256, activation='relu'))
# take any input and convert it into a 0 and 1
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam' , loss=tf.losses.BinaryCrossentropy() , metrics=['accuracy'])
# adam optimizers -> tf.optimizers.
```

```python
model.summary()
```

```python
logdir='logs'
```

```python
tensorboard_callback=tf.keras.callbacks.TensorBoard(log_dir=logdir)
# to save the model at a particulat checkpoint or for logging
```

```python
hist = model.fit(train, epochs=20 ,validation_data=val , callbacks=[tensorboard_callback])
```

```python
hist.history
```

```python
# plotting the performance
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal' , label='loss')
plt.plot(hist.history['val_loss'], color='orange' , label='val_loss')
fig.suptitle('Loss' , fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
# evaluate the accuracy
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal' , label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange' , label='val_accuracy')
fig.suptitle('Accuracy' , fontsize=20)
plt.legend(loc="upper left")
plt.show()
```
